```python
import BrainBERT.models as models


def build_model(cfg):
    ckpt_path = cfg.upstream_ckpt
    init_state = torch.load(ckpt_path, weights_only=False)
    upstream_cfg = init_state["model_cfg"]
    upstream = models.build_model(upstream_cfg)
    return upstream


def load_model_weights(model, states, multi_gpu):
    if multi_gpu:
        model.module.load_weights(states)
    else:
        model.load_weights(states)
```

```python
from omegaconf import OmegaConf
import torch

ckpt_path = "../dependencies/BrainBERT_weights/stft_large_pretrained.pth"
model = build_model(OmegaConf.create({"upstream_ckpt": ckpt_path}))
model.to('cuda')

init_state = torch.load(ckpt_path, weights_only=False)
load_model_weights(model, init_state['model'], False)
```

```python

import numpy as np
import os


# Remove metadata keys
data = {k: v for k, v in mat.items() if not k.startswith("__")}
os.makedirs("npy_output", exist_ok=True)

num_trials = data["joined_data"].shape[1]  # assuming same number for all

for i in range(num_trials):
    trial_data = {
        "joined_data": data["joined_data"][0, i],
        "labels_selfassessment": data["labels_selfassessment"][0, i],
        "labels_ext_annotation": data["labels_ext_annotation"][0, i],
        "VideoID": data["VideoIDs"][0, i] if "VideoIDs" in data else None,
    }
    # Save trial data as npz
    filename = f"npy_output/trial_{i}.npz"
    np.savez_compressed(filename, **trial_data)
```
